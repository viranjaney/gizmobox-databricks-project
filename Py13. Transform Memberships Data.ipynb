## Transform Memberships Data
1. Extract customer_id from the file path 
1. Write transformed data to the Silver schema  

```python
df_memberships = spark.read.table("gizmobox.bronze.py_memberships")
display(df_memberships)
```

```python
# 1.

from pyspark.sql import functions as F

df_extratced = (
    df_memberships
    .select(
        F.regexp_extract('path', '.*/([0-9]+)\\.png$', 1).alias('customer_id'),
        F.col('content').alias('membership_card')
    )
)
display(df_extratced)
```

```python
df_extratced.writeTo('gizmobox.silver.py_memberships').createOrReplace()
```

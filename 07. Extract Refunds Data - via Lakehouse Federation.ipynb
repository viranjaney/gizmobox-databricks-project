## Configure Access to Azure SQL Database via Lakehouse Federation using SQL
1. Create connection - asql_gizmobox_db_conn_sql
2. Create Foriegn Catalog - asql_gizmobox_db_catalog

```sql
-- use databricks secrets for secure connection
CREATE CONNECTION asql_gizmobox_db_conn_sql TYPE sqlserver
OPTIONS(
  host 'gizmobox-srv-1000.database.windows.net',
      port '1433',
      user 'gizmoboxadm',
      password '@vg83VG83@'
)
```

```sql
CREATE FOREIGN CATALOG IF NOT EXISTS asql_gizmobox_db_catalog USING CONNECTION asql_gizmobox_db_conn_sql
OPTIONS (database 'gizmobox-db');
```

```sql
SHOW CATALOGS;
USE CATALOG asql_gizmobox_db_catalog;
SHOW SCHEMAS;
USE SCHEMA dbo;
SHOW TABLES;
```

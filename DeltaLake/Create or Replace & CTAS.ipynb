1. Difference between Create or Replace and Drop and Create Table statements
2. CTAS statement




```sql
DROP TABLE IF EXISTS demo.delta_lake.companies;

CREATE TABLE demo.delta_lake.companies
  (
    company_id BIGINT NOT NULL GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
    company_name STRING NOT NULL,
    founded_date DATE COMMENT 'The date column',
    founded_year INT GENERATED ALWAYS AS (YEAR(founded_date)),
    country STRING
  );


INSERT INTO demo.delta_lake.companies (company_name, founded_date, country)
VALUES ('Microsoft', '1975-04-04', 'USA'),
       ('Google', '1998-09-04', 'USA'), 
      ('Amazon', '1994-05-05', 'USA')
```

```sql


CREATE OR REPLACE TABLE demo.delta_lake.companies
  (
    company_id BIGINT NOT NULL GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
    company_name STRING NOT NULL,
    founded_date DATE COMMENT 'The date column',
    founded_year INT GENERATED ALWAYS AS (YEAR(founded_date)),
    country STRING
  );


INSERT INTO demo.delta_lake.companies (company_name, founded_date, country)
VALUES ('Microsoft', '1975-04-04', 'USA'),
       ('Google', '1998-09-04', 'USA'), 
      ('TEK', '1994-05-05', 'USA'),
      ('META', '1998-09-04', 'USA'), 
      ('Amazon', '1994-05-05', 'USA')
```

```sql
CREATE TABLE demo.delta_lake.companies_usa
AS
SELECT * FROM demo.delta_lake.companies
WHERE country = 'USA'''
```

```sql
DESC HISTORY demo.delta_lake.companies_usa
```

```sql
CREATE OR REPLACE TABLE demo.delta_lake.companies_usa
AS
SELECT 
CAST(company_id AS INT) AS company_id,
company_name,
founded_year,
country
FROM demo.delta_lake.companies
WHERE country = 'USA'
```

```sql
ALTER TABLE demo.delta_lake.companies_usa
    ALTER COLUMN company_id SET NOT NULL
 
```

```sql
ALTER TABLE demo.delta_lake.companies_usa
   ALTER COLUMN founded_year COMMENT 'Date company was founded'
```

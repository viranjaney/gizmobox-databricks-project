# DLT pipeline renamed as Lakeflow declarative ETL pipeline

### Notes
1. We cannot mix-match SQL and python code in a single notebook for DLT. The DLT does not support it.
2. We can create seperate Python or SQL notebooks to run in a DLT pipeline.

Ex: We can write code in Pyspark for ingesting customers data in customers notebook and write SparkSQL code for ingesting orders data in orders notebook.

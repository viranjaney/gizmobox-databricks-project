## Extract Data From the Address Files
1. Demonstrate Limitations of CSV file format in SELECT statement
1. Use read_files function to overcome the limitations
1. Create Addresses view in the Bronze Layer

### 1. Demonstrate Limitations of CSV file format in SELECT statement

```python
df = spark.read.csv('/Volumes/gizmobox/landing/operational_data/addresses/')
display(df)
```

%md
### 2. Use read_files function to overcome the limitations
[Function Documentaion](https://learn.microsoft.com/en-us/azure/databricks/sql/language-manual/functions/read_files)

```python
df = spark.read \
    .format('csv') \
    .option('header', 'true') \
    .option('delimiter', '\t') \
    .load('/Volumes/gizmobox/landing/operational_data/addresses/')

display(df)
```

%md
### 3. Create Addresses table in the Bronze Layer

```python
df.writeTo('gizmobox.bronze.py_addresses').createOrReplace()
```

```python
%sql
select * from gizmobox.bronze.py_addresses;
```

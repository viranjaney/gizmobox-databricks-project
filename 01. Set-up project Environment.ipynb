## Set-up the project environment for GizmoBox Data Lakehouse

### 1. Access the container gizmobox

```sql
%python
ls 'abfss://gizmobox@deacourseexternaldl100.dfs.core.windows.net/landing'
```

### 2. Create external location

```sql
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS deacourseexternaldl100_gizmobox
URL 'abfss://gizmobox@deacourseexternaldl100.dfs.core.windows.net'
WITH (STORAGE CREDENTIAL dea_course_ext_sc )
COMMENT 'External location for gizmobox Lakehouse'
```

### 3. Create the catalog - gizmobox

```sql
%sql

SHOW CATALOGS;
```

```sql
%sql
 
 CREATE CATALOG  IF NOT EXISTS gizmobox
      MANAGED LOCATION 'abfss://gizmobox@deacourseexternaldl100.dfs.core.windows.net'
      COMMENT 'THis is the catalog for the gizmobox data lakehouse'
      
```

```sql
%sql

SHOW CATALOGS;
```

###4. Create Schemas
1. Landing
2. Bronze
3. Silver
4. Gold

```sql
%sql
select current_catalog();
```

```sql
use catalog gizmobox;

CREATE SCHEMA IF NOT EXISTS landing
    MANAGED LOCATION 'abfss://gizmobox@deacourseexternaldl100.dfs.core.windows.net/landing'
    COMMENT 'landing schema for gizmobox';
     
CREATE SCHEMA IF NOT EXISTS bronze
    MANAGED LOCATION 'abfss://gizmobox@deacourseexternaldl100.dfs.core.windows.net/bronze'
    COMMENT 'bronze schema for gizmobox';

CREATE SCHEMA IF NOT EXISTS silver
    MANAGED LOCATION 'abfss://gizmobox@deacourseexternaldl100.dfs.core.windows.net/silver'
    COMMENT 'silver schema for gizmobox';

CREATE SCHEMA IF NOT EXISTS gold
    MANAGED LOCATION 'abfss://gizmobox@deacourseexternaldl100.dfs.core.windows.net/gold'
    COMMENT 'gold schema for gizmobox';
```

```sql
%sql
SHOW SCHEMAS;
```

```sql
USE CATALOG gizmobox;
USE SCHEMA landing;

-- Volumes store unstructured, semi-structured and structured data
CREATE EXTERNAL VOLUME IF NOT EXISTS operational_data
    LOCATION 'abfss://gizmobox@deacourseexternaldl100.dfs.core.windows.net/landing/operational_data/'
    COMMENT 'Create external volume for operational data'
```

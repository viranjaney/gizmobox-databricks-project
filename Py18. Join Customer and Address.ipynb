## Join Customer and Address
Join customer data with address data to create a customer_address table which contains the address of each customer on the same record

```python
df_addresses = spark.read.table("gizmobox.silver.py_addresses")
df_customers = spark.read.table("gizmobox.silver.py_customers")
```

```python
df_customer_address = (
            df_addresses
            .join(df_customers, "customer_id", "inner")
            .select(
              "customer_id",
              "customer_name",
              "date_of_birth",
              "email",
              "member_since",
              "shipping_address_line_1",
              "shipping_city",
              "shipping_state",
              "shipping_postcode",
              "billing_address_line_1",
              "billing_city",
              "billing_state",
              "billing_postcode"
            )
)

display(df_customer_address)
```

```python
df_customer_address.writeTo("gizmobox.gold.py_customer_address").createOrReplace()
```
